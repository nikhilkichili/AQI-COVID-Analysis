```python
df_aqi_covid_2021_LA = df_aqi_covid_2021.loc[df_aqi_covid_2021['City'].isin(['Los Angeles'])]
df_aqi_covid_2021_LA = df_aqi_covid_2021_LA.drop(['City'], axis = 1)
df_aqi_covid_2021_HS = df_aqi_covid_2021.loc[df_aqi_covid_2021['City'].isin(['Houston'])]
df_aqi_covid_2021_HS = df_aqi_covid_2021_HS.drop(['City'], axis = 1)
df_aqi_covid_2021_DT = df_aqi_covid_2021.loc[df_aqi_covid_2021['City'].isin(['Detroit'])]
df_aqi_covid_2021_DT = df_aqi_covid_2021_DT.drop(['City'], axis = 1)
```

```python
plt.figure(figsize=(30,8))
sns.lineplot(data=df_aqi_covid_2021_LA.predicted_aqi, color="g")
ax2 = plt.twinx()
sns.lineplot(data=df_aqi_covid_2021_LA.cases, color="b", ax=ax2)
```

```python
#Create combo chart
fig, ax1 = plt.subplots(figsize=(30,8))
color = 'tab:green'
#bar plot creation
ax1.set_title('Covid and AQI for LA', fontsize=16)
ax1.set_xlabel('Date', fontsize=16)
ax1.set_ylabel('Cases', fontsize=16)
ax1 = sns.barplot(x='Date', y='cases', data = df_aqi_covid_2021_LA, palette='summer')
ax1.tick_params(axis='y')
#specify we want to share the same x-axis
ax2 = ax1.twinx()
color = 'tab:red'
#line plot creation
ax2.set_ylabel('AQI', fontsize=16)
ax2 = sns.lineplot(x='Date', y='predicted_aqi', data = df_aqi_covid_2021_LA, sort=False, color=color)
ax2.tick_params(axis='y', color=color)
#show plot
plt.show()
```
